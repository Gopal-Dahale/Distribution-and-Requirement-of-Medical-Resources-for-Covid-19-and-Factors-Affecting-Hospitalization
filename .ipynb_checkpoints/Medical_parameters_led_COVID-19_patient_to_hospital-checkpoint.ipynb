### Data Cleaning

The dataset contained data for both positive and negetive cases. But for our analysis I have to select the people those are tested positive.

```python
#Importing the clinical spectrum data
clinical_spectrum = pd.read_csv('dataset/diagnosis-of-covid-19-and-its-clinical-spectrum.csv')

#Filtering the data to contain the values only for the confirmed COVID-19 Tests
confirmed = clinical_spectrum['sars_cov_2_exam_result'] == 'positive'
clinical_spectrum = clinical_spectrum[confirmed]

#Printing total number of rows
print(len(clinical_spectrum))

#Viewing the dataset statisticsTo determine the parameters
clinical_spectrum.head()
```

From the result it seems like there are several columns that contains most of the NAN values. But I have went ahead with this dataset to see If I can find any parameter that led to covid-19 patient to hospital. 

### Making dataframe for analysis

**If patient need to admit in ICU (i.e. ICU required)**

```python
#Filetering the datasets
a_icu = clinical_spectrum['patient_addmited_to_intensive_care_unit_1_yes_0_no'] == 't'
a_icu = clinical_spectrum[a_icu]

na_icu = clinical_spectrum['patient_addmited_to_intensive_care_unit_1_yes_0_no'] == 'f'
na_icu = clinical_spectrum[na_icu]

#Taking mean and variance value of the spectra conditions
a_icu_mean = a_icu.mean(axis = 0, skipna = True) .to_frame().reset_index()
a_icu_var = a_icu.var(axis = 0, skipna = True) .to_frame().reset_index()

na_icu_mean = na_icu.mean(axis = 0, skipna = True) .to_frame().reset_index()
na_icu_var = na_icu.var(axis = 0, skipna = True) .to_frame().reset_index()

#Make the dataframe columns
a_icu_mean.columns = ['Parameter','a_icu_mean']
a_icu_var.columns = ['Parameter','a_icu_var']

na_icu_mean.columns = ['Parameter','na_icu_mean']
na_icu_var.columns = ['Parameter','na_icu_var']

#Merge dataframes
a_icu_mean['a_icu_var'] = a_icu_var['a_icu_var']
a_icu_mean['na_icu_mean'] = na_icu_mean['na_icu_mean']
a_icu_mean['na_icu_var'] = na_icu_var['na_icu_var']

#Assigning to a new name to dataframe for convenience
result = a_icu_mean

#Calculate mean difference
result['mean_diff'] = result['a_icu_mean']-result['na_icu_mean']

#View Dataframe
result.dropna().head()
```

```python
#Calculating scoring functon using above formula
c0 = 1
c1 = -0.2
c2 = -0.2
result['score'] = c0*result['mean_diff'] + c1*result['a_icu_var'] + c2*result['na_icu_var']

#Determining most important factor
result.sort_values(['score'], axis=0, ascending=False, inplace=True)
high_score = result.dropna().head(10)
high_score
```

```python
print('The difference between the hospitalised and non hospitalised patients can be\n somewhat made on the basic of the values of concentration of the following parameters:\n')
print('The values of these parameters vary the maximum between those who are hospitalised and who are not.')
print('Hence can be used to check for requirement of ICU\n')

for i  in list(high_score['Parameter']):
    print(i)
```

**If a patient's test result shows a_icu_mean values for the mentioned parameters with change nomore than mentioned variance, than he/she may required ICU.**

```python
ax = high_score.plot.bar(x='Parameter', y=['a_icu_mean','na_icu_mean'], rot=60,figsize=(10,5))
```

**If patient need to admit in hospital without ICU(i.e. normal beds requirement)**

```python
#Filetering the datasets
a_rward = clinical_spectrum['patient_addmited_to_regular_ward_1_yes_0_no'] == 't'
a_rward = clinical_spectrum[a_rward]

na_rward = clinical_spectrum['patient_addmited_to_regular_ward_1_yes_0_no'] == 'f'
na_rward = clinical_spectrum[na_rward]

#Taking mean and variance value of the spectra conditions
a_rward_mean = a_rward.mean(axis = 0, skipna = True) .to_frame().reset_index()
a_rward_var = a_rward.var(axis = 0, skipna = True) .to_frame().reset_index()

na_rward_mean = na_rward.mean(axis = 0, skipna = True) .to_frame().reset_index()
na_rward_var = na_rward.var(axis = 0, skipna = True) .to_frame().reset_index()

#Make the dataframe columns
a_rward_mean.columns = ['Parameter','a_rward_mean']
a_rward_var.columns = ['Parameter','a_rward_var']

na_rward_mean.columns = ['Parameter','na_rward_mean']
na_rward_var.columns = ['Parameter','na_rward_var']

#Merge dataframes
a_rward_mean['a_rward_var'] = a_rward_var['a_rward_var']
a_rward_mean['na_rward_mean'] = na_rward_mean['na_rward_mean']
a_rward_mean['na_rward_var'] = na_rward_var['na_rward_var']

#Assigning to a new name to dataframe for convenience
result = a_rward_mean

#Calculate mean difference
result['mean_diff'] = result['a_rward_mean']-result['na_rward_mean']

#View Dataframe
result.dropna().head()
```

I tried  a scoring function that is as follows: 
**score = c0\*mean_diff + c1\*a_rward_var + c2\*na_rward_var** \
I have tried different scoring coefficient. But the below mentioned one gives a decent result.

```python
#Calculating scoring functon using above formula
c0 = 1
c1 = -0.2
c2 = -0.2
result['score'] = c0*result['mean_diff'] + c1*result['a_rward_var'] + c2*result['na_rward_var']

#Determining most important factor
result.sort_values(['score'], axis=0, ascending=False, inplace=True)
high_score = result.dropna().head(10)
high_score
```

```python
print('The difference between the hospitalised and non hospitalised patients can be\n somewhat made on the basic of the values of concentration of the following parameters:\n')
print('The values of these parameters vary the maximum between those who are hospitalised and who are not.')
print('hence can be used to check for requirement of beds\n')

for i  in list(high_score['Parameter']):
    print(i)
```
